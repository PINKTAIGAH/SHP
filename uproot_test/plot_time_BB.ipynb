```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def linear(x,m,c):
    return m*x+c
```

```python
def find_rms(data):
    return np.sqrt(np.mean(data, axis=1)**2).flatten()
```

```python
P2H= np.loadtxt('../Data/P2H_rms_raw_data.txt')
BB= np.loadtxt('../Data/BB_rms_raw_data.txt')
```

```python
P2H_long= np.loadtxt('../Data/P2H_rms_raw_data_long.txt')
BB_long= np.loadtxt('../Data/BB_rms_raw_data_long.txt')
```

```python
P2H_timeav, P2H_x,P2H_errtime= (P2H[0].reshape(7,20), P2H[1].reshape(7,20), P2H[2].reshape(7,20))
BB_timeav, BB_x,BB_errtime= (BB[0].reshape(7,20), BB[1].reshape(7,20), BB[2].reshape(7,20))
```

```python
P2H_timeav_long, P2H_x_long,P2H_errtime_long= (P2H_long[0].reshape(7,20), P2H_long[1].reshape(7,20), P2H_long[2].reshape(7,20))
BB_timeav_long, BB_x_long,BB_errtime_long= (BB_long[0].reshape(7,20), BB_long[1].reshape(7,20), BB_long[2].reshape(7,20))
```

```python
P2H_rms= find_rms(P2H_timeav)
P2H_xaxis_rms= P2H_x.T[0]

BB_rms= find_rms(BB_timeav)
BB_xaxis_rms= BB_x.T[0]
```

```python
P2H_rms_long= find_rms(P2H_timeav_long)
P2H_xaxis_rms_long= P2H_x_long.T[0]

BB_rms_long= find_rms(BB_timeav_long)
BB_xaxis_rms_long= BB_x_long.T[0]
```

```python
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(P2H_x.flatten(), P2H_timeav.flatten(), yerr=P2H_errtime.flatten(), fmt='bx', ecolor='k', markersize= 1.5, label='Project to Hight')
ax1.set_ylim((1.8,1.9))
ax1.legend(loc='lower left')
ax1.set_xlabel('Prioject to height, P2H (cm)')
ax1.set_ylabel(r'runtime, t$_{gen}$ (s)')

ax2.errorbar(BB_x.flatten(), BB_timeav.flatten(), yerr=BB_errtime.flatten(), fmt='rx', ecolor='k', markersize= 1.5, label= 'buffer Box')
ax2.set_ylim((1.8,1.9))
ax2.legend(loc='lower left')
ax2.set_xlabel('buffer box, BB (cm)')
ax2.set_ylabel(r'runtime, t$_{gen}$ (s)')
ax1.set_title('Cosmic generation runtime scatter plot', y=1.05)
fig.tight_layout()
```

```python
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(P2H_xaxis_rms, np.average(P2H_timeav, axis=1), yerr= np.std(P2H_timeav, axis=1)/np.sqrt(P2H_timeav.shape[1]), fmt= 'bx',ecolor='k', markersize= 6, label= 'Project to Hight')
ax1.legend(loc='lower right')

ax2.errorbar(BB_xaxis_rms, np.average(BB_timeav, axis=1),yerr= np.std(BB_timeav, axis=1)/np.sqrt(BB_timeav.shape[1]),fmt= 'rx', ecolor='k', markersize= 6, label= 'Buffer Box')
ax2.legend(loc='lower right')

ax1.set_xlabel('Prioject to height, P2H (cm)')
ax1.set_ylabel(r'average runtime, t$_{gen}$ (s)')
ax2.set_xlabel('buffer box, BB (cm)')
ax2.set_ylabel(r'average runtime, t$_{gen}$ (s)')
ax1.set_title('Average runtime of cosmic generation ', y=1.05)
fig.tight_layout()
```

```python
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(P2H_x_long.flatten(), P2H_timeav_long.flatten(), yerr=P2H_errtime_long.flatten(), fmt='bx', ecolor='k', markersize= 1.5, label='Project to Hight')
ax1.set_ylim((1.8,1.9))
ax1.legend(loc='lower left')
ax1.set_xlabel('Prioject to height, P2H (cm)')
ax1.set_ylabel(r'runtime, t$_{gen}$ (s)')

ax2.errorbar(BB_x_long.flatten(), BB_timeav_long.flatten(), yerr=BB_errtime_long.flatten(), fmt='rx', ecolor='k', markersize= 1.5, label= 'buffer Box')
ax2.set_ylim((1.8,1.9))
ax2.legend(loc='lower left')
ax2.set_xlabel('buffer box, BB (cm)')
ax2.set_ylabel(r'runtime, t$_{gen}$ (s)')
ax1.set_title('Cosmic generation runtime scatter plot', y=1.05)
fig.tight_layout()
```

```python
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(P2H_xaxis_rms, np.average(P2H_timeav_long, axis=1), yerr= np.std(P2H_timeav_long, axis=1)/np.sqrt(P2H_timeav.shape[1]), fmt= 'bx',ecolor='k', markersize= 6, label= 'Project to Hight')
ax1.legend(loc='lower right')

ax2.errorbar(BB_xaxis_rms, np.average(BB_timeav_long, axis=1),yerr= np.std(BB_timeav_long, axis=1)/np.sqrt(BB_timeav.shape[1]),fmt= 'rx', ecolor='k', markersize= 6, label= 'Buffer Box')
ax2.legend(loc='lower right')

ax1.set_xlabel('Prioject to height, P2H (cm)')
ax1.set_ylabel(r'average runtime, t$_{gen}$ (s)')
ax2.set_xlabel('buffer box, BB (cm)')
ax2.set_ylabel(r'average runtime, t$_{gen}$ (s)')
ax1.set_title('Average runtime of cosmic generation ', y=1.05)
fig.tight_layout()
```

```python
gen_long= np.loadtxt('../Data/BB_gen_long_raw_data.txt')
gen_short= np.loadtxt('../Data/BB_gen_short_raw_data.txt')

gen_timeav_long, gen_x_long,gen_errtime_long= (gen_long[0].reshape(14,10), gen_long[1].reshape(14,10), gen_long[2].reshape(14,10))
gen_timeav_short, gen_x_short,gen_errtime_short= (gen_short[0].reshape(14,10), gen_short[1].reshape(14,10), gen_short[2].reshape(14,10))

gen_errtime_long[np.where(gen_timeav_long<1.5)] = 0
gen_timeav_long[gen_timeav_long<1.5] = 1.8


fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(gen_x_long.flatten(), gen_timeav_long.flatten(), yerr=gen_errtime_long.flatten(), fmt='bx', ecolor='k', markersize= 3, label='Buffer Box')
# ax1.set_ylim((1.5,2.3))
ax1.legend(loc='upper right')
ax1.set_xlabel('buffer box length, BB (cm)')
ax1.set_ylabel(r'runtime, t$_{gen}$ (s)')

ax2.errorbar(np.average(gen_x_long, axis= 1).flatten(), np.average(gen_timeav_long, axis=1),yerr= np.std(gen_timeav_long, axis=1)/np.sqrt(gen_timeav_long.shape[1]),fmt= 'rx', ecolor='k', markersize= 6, label= 'Buffer Box average')
ax2.legend(loc='upper right')
ax2.set_xlabel('buffer box length, BB (cm)')
ax2.set_ylabel(r'average runtime, t$_{gen}$ (s)')
# ax1.set_title('Event runtime vs buffer box size for cosmic generation', y=1.05)
fig.tight_layout()


corr_long= stats.pearsonr(np.average(gen_timeav_long, axis=1).flatten(), np.average(gen_x_long, axis= 1).flatten())
print(f'The pearson statistic for long is {corr_long.statistic:.3} with a p value of {corr_long.pvalue:.3}')
```

```python
gen_long= np.loadtxt('../Data/BB_gen_long_raw_data.txt')
gen_short= np.loadtxt('../Data/BB_gen_short_raw_data.txt')

gen_timeav_long, gen_x_long,gen_errtime_long= (gen_long[0].reshape(14,10), gen_long[1].reshape(14,10), gen_long[2].reshape(14,10))
gen_timeav_short, gen_x_short,gen_errtime_short= (gen_short[0].reshape(14,10), gen_short[1].reshape(14,10), gen_short[2].reshape(14,10))

gen_errtime_short[np.where(gen_timeav_short<1.5)] = 0
gen_timeav_short[gen_timeav_short<1.5] = 1.8


fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(gen_x_short.flatten(), gen_timeav_short.flatten(), yerr=gen_errtime_short.flatten(), fmt='bx', ecolor='k', markersize= 5, label='Buffer Box')
# ax1.set_ylim((1.5,2.3))
ax1.legend(loc='upper left')
ax1.set_xlabel('buffer box length, BB (cm)')
ax1.set_ylabel(r'runtime, t$_{gen}$ (s)')

ax2.errorbar(np.average(gen_x_short, axis= 1).flatten(), np.average(gen_timeav_short, axis=1),yerr= np.std(gen_timeav_short, axis=1)/np.sqrt(gen_timeav_short.shape[1]),fmt= 'rx', ecolor='k', markersize= 6, label= 'Buffer Box average')
ax2.legend(loc='upper right')
ax2.set_xlabel('buffer box length, BB (cm)')
ax2.set_ylabel(r'average runtime, t$_{gen}$ (s)')
ax1.set_title('Event runtime vs buffer box size for cosmic generation', y=1.05)
fig.tight_layout()


corr_short= stats.pearsonr(np.average(gen_timeav_short, axis= 1).flatten(), np.average(gen_x_short, axis= 1).flatten())
print(f'The pearson statistic for short is {corr_short.statistic:.3} with a p value of {corr_short.pvalue:.3}')
```

```python
g4_long= np.loadtxt('../Data/BB_g4_long_raw_data.txt')
g4_short= np.loadtxt('../Data/BB_g4_short_raw_data.txt')

g4_timeav_long, g4_x_long,g4_errtime_long= (g4_long[0].reshape(14,10), g4_long[1].reshape(14,10), g4_long[2].reshape(14,10))
g4_timeav_short, g4_x_short,g4_errtime_short= (g4_short[0].reshape(14,10), g4_short[1].reshape(14,10), g4_short[2].reshape(14,10))

reg= stats.linregress(np.average(g4_x_long, axis= 1).flatten(), np.average(g4_timeav_long, axis=1).flatten())
print(f'The pearson statistic for long is {reg.rvalue:.3} with a p value of {reg.pvalue:.3}')
print(f'The linear regression for long returns a slope of {reg.slope:.3} *- {reg.stderr} and a intercept of {reg.intercept:.3} +- {reg.intercept_stderr}')
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(g4_x_long.flatten(), g4_timeav_long.flatten(), yerr=g4_errtime_long.flatten(), fmt='bx', ecolor='k', markersize= 3, label='Buffer Box')
# ax1.set_ylim((1.5,2.3))
ax1.legend(loc='upper left')
ax1.set_xlabel('buffer box length, BB (cm)')
ax1.set_ylabel(r'runtime, t$_{gen}$ (s)')

ax2.errorbar(np.average(g4_x_long, axis= 1).flatten(), np.average(g4_timeav_long, axis=1),yerr= np.std(g4_timeav_long, axis=1)/np.sqrt(g4_timeav_long.shape[1]),fmt= 'rx', ecolor='k', markersize= 6, label= 'Buffer Box average')
ax2.set_xlabel('buffer box length, BB (cm)')
ax2.set_ylabel(r'average runtime, t$_{gen}$ (s)')
ax2.errorbar(np.arange(g4_x_long[0][0], g4_x_long[-1][0]), linear(np.arange(g4_x_long[0][0], g4_x_long[-1][0]), reg.slope, reg.intercept),fmt='k--', label= 'Fitted line')
ax2.legend(loc='upper left')
# ax1.set_title('Event runtime vs buffer box size for geant4 simulation', y=1.05)
fig.tight_layout()
```

```python
anomaly_mask= [True, True,True, True,True, True,True, True,True, False,True, True,True, True,]
x= g4_x_long[:,1]
residual= linear(x, reg.slope, reg.intercept)- np.average(g4_timeav_long, axis=1)
residual_err= np.std(g4_timeav_long, axis=1)/np.sqrt(g4_timeav_long.shape[1])
chi_2= (residual[anomaly_mask]**2/residual_err[anomaly_mask]**2).sum()
print(f'Chi squared is {chi_2}')

plt.errorbar(x, residual, yerr=residual_err, fmt= 'rx', ecolor='k', markersize= 7, label= 'Buffer Box residuals')
plt.legend(loc='upper left')
plt.xlabel('buffer box length, BB (cm)')
plt.ylabel(r'residuals (s)')
plt.axhline(c='k', alpha=0.5)
# plt.title('Residuals of event runtime vs buffer box size for geant4 simulation', y=1.05)
```

```python
det_long= np.loadtxt('../Data/BB_det_long_raw_data.txt')
det_short= np.loadtxt('../Data/BB_det_short_raw_data.txt')

det_timeav_long, det_x_long,det_errtime_long= (det_long[0].reshape(14,10), det_long[1].reshape(14,10), det_long[2].reshape(14,10))
det_timeav_short, det_x_short,det_errtime_short= (det_short[0].reshape(14,10), det_short[1].reshape(14,10), det_short[2].reshape(14,10))

det_errtime_long[np.where(det_timeav_long>30)] = 0
det_timeav_long[det_timeav_long>30] = 20

reg= stats.linregress(np.average(det_x_long, axis= 1).flatten(), np.average(det_timeav_long, axis=1).flatten())
print(f'The pearson statistic for long is {reg.rvalue:.3} with a p value of {reg.pvalue:.3}')
print(f'The linear regression for long returns a slope of {reg.slope:.3} *- {reg.stderr} and a intercept of {reg.intercept:.3} +- {reg.intercept_stderr}')
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.errorbar(det_x_long.flatten(), det_timeav_long.flatten(), yerr=det_errtime_long.flatten(), fmt='bx', ecolor='k', markersize= 5, label='Buffer Box')
# ax1.set_ylim((1.5,2.3))
ax1.legend(loc='upper center')
ax1.set_xlabel('buffer box length, BB (cm)')
ax1.set_ylabel(r'runtime, t$_{gen}$ (s)')

ax2.errorbar(np.average(det_x_long, axis= 1).flatten(), np.average(det_timeav_long, axis=1),yerr= np.std(det_timeav_long, axis=1)/np.sqrt(det_timeav_long.shape[1]),fmt= 'rx', ecolor='k', markersize= 6, label= 'Buffer Box average')
ax2.set_xlabel('buffer box length, BB (cm)')
ax2.set_ylabel(r'average runtime, t$_{gen}$ (s)')
# ax2.errorbar(np.arange(det_x_long[0][0], det_x_long[-1][0]), linear(np.arange(det_x_long[0][0], det_x_long[-1][0]), reg.slope, reg.intercept),fmt='k--', label= 'linear regression')
ax2.legend(loc='lower left')
# ax1.set_title('Event runtime vs buffer box size for detector simulation', y=1.05)
fig.tight_layout()
```
